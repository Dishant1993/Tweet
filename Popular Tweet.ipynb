# Promote brand around popular retweets


1.You are creating branding message for one of your fitness brands

2.Identify popular tweets on fitness that have been retweeted many times to bundle your branding message

```python
# Import libraries
import pandas as pd
import numpy as np
```

```python
# Import extracted tweets on IoT
tweet_fit  = pd.read_csv("tweet_fitness.csv")
```

```python
# View basic info about the dataset
tweet_fit.info()
```

```python
# View dataset
tweet_fit.head()
```

```python
# Create a subset dataframe of tweet text and retweet count
rtwt = tweet_fit[['text', 'retweet_count']]
rtwt
```

```python
# Print rtwt with two columns
rtwt.head()
```

```python
# Sort dataframe based on descending order of retweet counts
rtwt1 = rtwt.sort_values(by='retweet_count', ascending=False)
```

```python
# Remove duplicate rows, keeping only once instance
rtwt2 = rtwt1.drop_duplicates(subset='text', keep='first')
```
